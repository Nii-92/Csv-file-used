```python
import pandas as pd

df = pd.read_csv("gender_submission.csv")

df.head()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()

for index, row in df.iterrows():
    user = row["PassengerId"]
    followers = row["Survived"]
    G.add_edge(user, followers)



plt.figure(figsize=(15, 15))
nx.draw(G, with_labels=False, node_color="red", edge_color="gray", node_size=50)
plt.title("gender_submission")
plt.show()
```

```python
centrality = nx.degree_centrality(G)
passenger_by_survival = {}


for passenger, score in centrality.items():

    survival_status_series = df.loc[df['PassengerId'] == passenger, 'Survived']

    if not survival_status_series.empty:
        survival_status = survival_status_series.iloc[0]
        if survival_status not in passenger_by_survival:
            passenger_by_survival[survival_status] = []
        passenger_by_survival[survival_status].append((passenger, int(score * (len(G) - 1))))


for survival_status, passengers in passenger_by_survival.items():
    print(f"\nPassengers with Survival Status {survival_status}:")
    passengers.sort(key=lambda x: x[1], reverse=True)
    for passenger, connections in passengers[:5]:
        print(f"  Passenger {passenger}: Connections - {connections}")
```

```python
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("gender_submission.csv")

G = nx.Graph()

for index, row in df.iterrows():
    user = row["PassengerId"]
    followers = row["Survived"]
    G.add_edge(user, followers)

betweenness_centrality = nx.betweenness_centrality(G)

passenger_by_survival = {}

for passenger, score in betweenness_centrality.items():
    survival_status_series = df.loc[df['PassengerId'] == passenger, 'Survived']
    if not survival_status_series.empty:
        survival_status = survival_status_series.iloc[0]
        if survival_status not in passenger_by_survival:
            passenger_by_survival[survival_status] = []
        passenger_by_survival[survival_status].append((passenger, score))


for survival_status, passengers in passenger_by_survival.items():
    print(f"\nPassengers with Survival Status {survival_status}:")
    passengers.sort(key=lambda x: x[1], reverse=True)
    for passenger, betweenness in passengers[:5]:
        print(f"  Passenger {passenger}: Betweenness Centrality - {betweenness}")
```

```python
clustering_coefficients = nx.clustering(G)

passenger_by_survival = {}

for passenger, score in clustering_coefficients.items():
    survival_status_series = df.loc[df['PassengerId'] == passenger, 'Survived']
    if not survival_status_series.empty:
        survival_status = survival_status_series.iloc[0]
        if survival_status not in passenger_by_survival:
            passenger_by_survival[survival_status] = []
        passenger_by_survival[survival_status].append((passenger, score))


for survival_status, passengers in passenger_by_survival.items():
    print(f"\nPassengers with Survival Status {survival_status}:")
    passengers.sort(key=lambda x: x[1], reverse=True)
    for passenger, clustering_coefficient in passengers[:5]:
        print(f"  Passenger {passenger}: Clustering Coefficient - {clustering_coefficient}")
```

```python

from google.colab import files
import pandas as pd
import networkx as nx

uploaded = files.upload()

df = pd.read_csv("gender_submission.csv")

G = nx.Graph()
for _, row in df.iterrows():
    G.add_node(row['PassengerId'], survived=row['Survived'])

connected_components = list(nx.connected_components(G))

print("Connected Components:")
for i, component in enumerate(connected_components):
    print(f"Component {i+1}: {component}")

```

```python
ER_graph = nx.erdos_renyi_graph(G.number_of_nodes(), G.number_of_edges() / (G.number_of_nodes() - 1))

degree_distribution_ER = [d[1] for d in ER_graph.degree()]
print(f"Average Degree of ER Graph: {sum(degree_distribution_ER) / len(degree_distribution_ER)}")
```

```python
n = G.number_of_nodes()
m = int(G.number_of_edges() / G.number_of_nodes())
m = max(1, min(m, n - 1))  # Ensure 1 <= m < n


BA_graph = nx.barabasi_albert_graph(n, m)

degree_distribution_BA = [d[1] for d in BA_graph.degree()]
print(f"Average Degree of BA Graph: {sum(degree_distribution_BA) / len(degree_distribution_BA)}")

```

```python
WS_graph = nx.watts_strogatz_graph(G.number_of_nodes(), 4, 0.1)

degree_distribution_WS = [d[1] for d in WS_graph.degree()]
print(f"Average Degree of WS Graph: {sum(degree_distribution_WS) / len(degree_distribution_WS)}")
```

```python
nx.write_gexf(G, "gender_submission.gexf")
```
