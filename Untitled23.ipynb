```python
expected_columns = ['State_Code', 'Category', 'Brand', 'Status']
available_columns = [col for col in expected_columns if col in df.columns]
df = pd.get_dummies(df, columns=available_columns, drop_first=True)
print("\n## One-hot encoding applied.")
```

```python
numeric_df = df.select_dtypes(include=['number'])
plt.figure(figsize=(18, 6))
corr_matrix = numeric_df.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('### Correlation Matrix of Numerical Features')
plt.show()
```

```python
plt.figure(figsize=(18, 5))
sns.histplot(df['Sales'], bins=30, kde=True)
plt.title('### Distribution of Sales')
plt.xlabel('Sales Amount')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(18, 6))
sns.scatterplot(x=df['Cost'], y=df['Sales'])
plt.title('### Correlation between Cost and Sales')
plt.xlabel('Cost')
plt.ylabel('Sales')
plt.show()
```
